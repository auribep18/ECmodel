```python
import statsmodels.api as sm

# Get the unique pollutants
pollutants = filtered_df['PrimaryPollutant'].unique()

# Dictionary to store the regression results for each pollutant
results = {}

# Loop over each pollutant and run the regression
for pollutant in pollutants:
    # Subset the data for the current pollutant
    subset = filtered_df[filtered_df['PrimaryPollutant'] == pollutant]
    
    # Prepare the data for regression
    X = subset[['Year', 'Emission_Gg']]
    y = subset['Value_ngperm3_per1Gg']
    
    # Add a constant to the independent variables matrix
    X = sm.add_constant(X)
    
    # Fit the regression model
    model = sm.OLS(y, X).fit()
    
    # Store the results
    results[pollutant] = model

    # Print the summary for the current pollutant
    print(f"Regression Results for {pollutant}")
    print(model.summary())
    print("\n" + "="*80 + "\n")
```
